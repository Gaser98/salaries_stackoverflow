```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt
```

```python
stackOverflow = pd.read_csv("survey_results_public.csv")
```

```python
stackOverflow.head()
```

```python
stackOverflow.describe()
```

```python
stackOverflow['StackOverflowSatisfaction']
```

```python
stackOverflow['StackOverflowSatisfaction'].fillna(0,inplace=True)
stackOverflow['Salary'].fillna(0,inplace=True)
```

```python
StackOverflowSatisfaction=stackOverflow['StackOverflowSatisfaction']
Salary=stackOverflow['Salary']
```

```python
x=Salary[:,np.newaxis]
y=StackOverflowSatisfaction
reg = LinearRegression()
reg.fit(x,y)
y_pred = reg.predict(x)
plt.scatter(x, y)
plt.plot(x, y_pred, color='green', linewidth=4)
```

```python
#As salaries increase,the satisfaction at the working place incrrases
#No patterns or outliers shown which depicts the right choice of the model
```

```python
residual = y - y_pred
plt.scatter(x,residual)
```
